```python
#Importing Necessary libraries (Lasio for .LAS files)
import lasio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#Creating empty list to append with API values
API_List = list()    
```

```python
#os.listdir for the current working directory, 
# be sure the las files are IN the working
# directory or this will not work
# Here we're iterating over each filename in the For loop
# This will search each file for the listed API number/well number, 
# and attempts both API and UWI values, whichever may
# be stored. This WILL differ from file to file, and some 
# have bizarre naming conventions.

# It appends the list with each filename, API number and whatever is listed in the 'LOC' or location
# section of each file. Some have useful info, some do not, as seen in the resultant .csv file

# Note: Keep the exceptions in place. There are a huge number of errors that will arise from reading these
#  highly variable log files, and not passing on exceptions will kick up errors every time.
for filename in os.listdir(os.getcwd()):
    if filename.endswith(".LAS") or filename.endswith('.las'): #Include .LAS and .las files
        try:
            well_data = lasio.read(filename)
            API_List.append((filename, 
                             well_data.well.API.value, 
                             well_data.well.LOC.value, 
                             well_data.well.STRT.value, 
                             well_data.well.STOP.value))
            continue
        except Exception:
            try:
                well_data = lasio.read(filename)
                API_List.append((filename, 
                                 well_data.well.UWI.value, 
                                 well_data.well.LOC.value, 
                                 well_data.well.STRT.value, 
                                 well_data.well.STOP.value))
            except Exception:
                pass
    else:
        continue
```

```python
#Display the first 10 values of API list for verification
API_List[0:9]
```

```python
# Convert the list into a pandas DataFrame 
# for easier manipulation and sorting/addition/summarizations
API_Dataframe = pd.DataFrame(API_List, 
                             columns = 
                             ['Filename', 'API', 'LOC', 'STRT DEPT', 'TVD'])
```

```python
#Display first few pieces of Dataframe for verification
API_Dataframe.head(10)
```

```python
#Bread and butter of the code...
#Defines the function 'curve_search', takes 2 arguments:
#    1 - A LIST of mnemonics/acronyms to search each listed log file for. 
#      - It will append the dataframe with each 
#        one, inputing a value of 1 if it contains, 0 if not
#    2 - Second argument is the dataframe to include. 
#      - The dataframe needs to have one column named 'Filename' that
#        containts the list of log filenames to search. 
#      - This function will alter and return the dataframe. Your original
#        will be lost. 
#      - Need to alter this to make a copy rather than write over the dataframe.
def curve_search(list_x, APIs):
    for x in list_x:
        x_list = []
        for API in APIs['Filename']:
            x_df = lasio.read(API).df()
            if x in x_df.columns:
                x_list.append(1)
            else:
                x_list.append(0)
        APIs[x] = x_list
    return APIs
```

```python
#Sample of calling the function with arguments in place
check = curve_search(['GR', 
                      'PE', 
                      'NPHI', 
                      'DPHI', 
                      'XPHI', 
                      'SPHI', 
                      'LLD', 
                      'MSFL', 
                      'RHOB', 
                      'POTA'], 
                     API_Dataframe)
```

```python
#Check the first few values of the newly created dataframe. 
check.head()
```

```python
#Export the dataframe to a CSV file
# Include the index/header!
check.to_csv(r'export_dataframe_1.csv', 
             index = True, 
             header= True)
```
