```python
# Import necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
```

```python
#import dataframe with API list to scrape
crockett_list = pd.read_csv('Crockett_Corrected_API.csv')
crockett_list = crockett_list.drop(['Unnamed: 0', 
                          'Unnamed: 0.1'], axis = 1)
crockett_list
```

```python
reagan_list = pd.read_csv('Reagan_Corrected_API_Values.csv')
reagan_list = reagan_list.drop(['Unnamed: 0', 
                                'Unnamed: 0.1'], axis = 1)
reagan_list
```

```python
def web_scrape_REAGAN(list_of_apis):
# Define initial dataframe and columns (for REAGAN COUNTY, ELSE THIS WILL CHANGE)
    df = pd.DataFrame(columns = ['API', 
                                 'Website', 
                                 'Block', 
                                 'Section', 
                                 'Latitude', 
                                 'Longitude', 
                                 'Field', 
                                 'Operator'])
# define base url to be added to
    url_frontend = 'http://www.utlands.utsystem.edu/API/'
    for api in list_of_apis:
        full_url = url_frontend + str(api)
# create list instance, first two values of dataframe = api number and url (will use api later to merge dataframes)
        ls = [api, full_url]
# Beautiful Soup get requests > text grabbing        
        page = requests.get(full_url)
        soup = bs(page.content, 'html.parser')
#Some ridiculously ugly code, let's see if it accomplishes the task
        soupstring = str(soup.find_all('tr')[0:1])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[19:21], cleantext[31:33]])
        soupstring = str(soup.find_all('tr')[1:2])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[30:41], cleantext[44:57]])
        if ls[-2] != '':
            if ls[-2][0] == ':':
                ls[-2] = cleantext[31:42]
                ls[-1] = cleantext[45:58]
        else:
            continue
        soupstring = str(soup.find_all('tr')[3:4])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[7:]])
        soupstring = str(soup.find_all('tr')[4:5])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[10:]])
        df.loc[len(df)] = ls     
    return df
```

```python
def web_scrape_CROCKETT(list_of_apis):
# Define initial dataframe and columns (for REAGAN COUNTY, ELSE THIS WILL CHANGE)
    df = pd.DataFrame(columns = ['API', 
                                 'Website', 
                                 'Block', 
                                 'Section', 
                                 'Latitude', 
                                 'Longitude', 
                                 'Field', 
                                 'Operator'])
# define base url to be added to
    url_frontend = 'http://www.utlands.utsystem.edu/API/'
    for api in list_of_apis:
        full_url = url_frontend + str(api)
# create list instance, first two values of dataframe = api number and url (will use api later to merge dataframes)
        ls = [api, full_url]
# Beautiful Soup get requests > text grabbing        
        page = requests.get(full_url)
        soup = bs(page.content, 'html.parser')
#Some ridiculously ugly code, let's see if it accomplishes the task
        soupstring = str(soup.find_all('tr')[0:1])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[21:23], cleantext[33:35]])
        soupstring = str(soup.find_all('tr')[1:2])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[30:41], cleantext[44:57]])
        if ls[-2] != '':
            if ls[-2][0] == ':':
                ls[-2] = cleantext[31:42]
                ls[-1] = cleantext[45:58]
        else:
            continue
        soupstring = str(soup.find_all('tr')[3:4])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[7:]])
        soupstring = str(soup.find_all('tr')[4:5])
        cleantext = bs(soupstring, 'html').get_text().replace('\n', '')
        ls.extend([cleantext[10:]])
        df.loc[len(df)] = ls     
    return df
```

```python
reagan_scraped = web_scrape_REAGAN(reagan_list['API'])
crockett_scraped = web_scrape_CROCKETT(crockett_list['API'])
```

```python
print(reagan_scraped.head(10))
print(crockett_scraped.head(10))
```

```python
# merging original dataframes with log info and new dataframes with website info
new_reagan = pd.merge(reagan_list, reagan_scraped, on = 'API')
new_crockett = pd.merge(crockett_list, crocket_scraped, on = 'API')
```

```python
#create a list in the desired order for the columns of new dataframe
column_order = ['API', 
                 'Website', 
                 'Section', 
                 'Block', 
                 'Latitude', 
                 'Longitude', 
                 'Operator', 
                 'Field', 
                 'GR', 
                 'LLD', 
                 'MSFL', 
                 'DPHI', 
                 'NPHI', 
                 'PE', 
                 'XPHI', 
                 'SPHI',
                 'RHOB',
                 'POTA', 
                 'Filename']
```

```python
#create the final dataframe, will columns in correct order, original LOC dropped (las file LOC data was terrible)
reagan_final = new_reagan[column_order]
crockett_final = new_crockett[column_order]
```

```python
reagan_final = reagan_final.drop_duplicates()
crockett_final = crockett_final.drop_duplicates()
```

```python
reagan_final.to_csv(r'completed_Reagan_county.csv', 
             index = True, 
             header= True)
crockett_final.to_csv(r'completed_Crockett_county.csv', 
             index = True, 
             header= True)
```
